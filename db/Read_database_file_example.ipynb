```python
# Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine, inspect, func
```

```python
# Create Engine for census data
project2_database_path = "Project2.db"
engine = create_engine(f"sqlite:///{project2_database_path}")
conn = engine.connect()
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
divvy_stationinfo_data = pd.read_sql("SELECT * FROM divvy_stationinfo", conn)
divvy_stationinfo_data.head()
```

```python
divvy_stationstatus_data = pd.read_sql("SELECT * FROM divvy_stationstatus", conn)
divvy_stationstatus_data.head()
```

```python
combined_divvy_data = pd.merge(divvy_stationinfo_data, divvy_stationinfo_data, on="station_id", how="inner")
combined_divvy_data
```
